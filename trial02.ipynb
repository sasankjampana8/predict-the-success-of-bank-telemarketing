Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
import xgboost as xg
parameter_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'gamma': [0, 0.1],
    'min_child_weight': [1, 3]
}


model = xg.XGBClassifier()
model_gridsearch = GridSearchCV(estimator=model, param_grid=parameter_grid, cv=5,  n_jobs=-1, scoring="accuracy")
model_gridsearch.fit(X_train, Y_train)
```

```python
# Print the best estimator and best score
print("Best Estimator:", model_gridsearch.best_estimator_)
print("Best Score:", model_gridsearch.best_score_)
```

```python
import xgboost as xg
model = xg.XGBClassifier()
model.fit(X_train, Y_train)
predictions = model.predict(X_val)
```

```python
predictions_best = model_gridsearch.best_estimator_.predict(X_val)
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_val, predictions, digits=4))
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_val, predictions_best, digits=4))
```
